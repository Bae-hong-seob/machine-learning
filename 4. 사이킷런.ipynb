```python
# sklearn에서 제공하는 붓꽃 데이터세트 다운받기
from sklearn.datasets import load_iris
# 머신러닝 알고리즘으로 의사 결정 트리(Decision Tree) 알고리즘 사용
from sklearn.tree import DecisionTreeClassifier
# 데이터를 학습데이터 세트 / 테스트 데이터 세트로 나눈다 -> train_test_split()함수 사용
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

# 붓꽃 데이터 세트를 로딩합니다.
iris = load_iris()

# iris.data는 Iris 데이터 세트에서 피처(feature)만으로 된 데이터를 numpy로 가지고 있습니다.
iris_data = iris.data

# iris.target은 붓꽃 데이터 세트에서 레이블(결정 값) 데이터를 numpy로 가지고 있습니다.
iris_label = iris.target
print('iris target값:', iris_label)
print('iris target명:', iris.target_names)

# 붓꽃 데이터 세트를 자세히 보기 위해 DataFrame으로 변환합니다.
iris_df = pd.DataFrame(data=iris_data, columns = iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

```python
# 이 붓꽃 학습 알고리즘
# 1. 붓꽃 사진을 input으로 받는다
# 2. 붓꽃을 특징을 가지고 있다. -> Sepal length, Sepal width, Petal length, Petal width
# 3. 그에 따라 붓꽃을 나눈다.(정답)) -> Setosa, Vesicolor, Virginica 세 종류 중 하나

# 학습 과정
# 1. 붓꽃의 특징과 정답(label)을 데이터 세트만큼 입력받는다. 
#     label 0은 Setosa, 1은 Versicolor, 2는 Virginica
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, 
                                                    iris_label, test_size = 0.2, 
                                                    random_state = 11)
```

```python
#iris_data
print(len(iris_data)) # 데이터 세트의 feature를 numpy 형식으로 저장된 값
print(len(iris_label)) # 데이터 세트의 label를 numpy 형식으로 저장된 값
```

```python
# train_test_split() 함수를 사용하여 데이터 세트를 학습 데이터 세트 / 테스트 데이터 세트로 나눈다.
# 첫번째 parameter로 데이터 세트 feature 값을
# 두번째 parameter로 데이터 세트 중 정답(label) 값을
# 세번째 parameter로 테스트 데이터 세트의 비율을
    # test_size = 0.2 -> 테스트 데이터 세트는 20%를 차지한다.
# 네번째 parameter로 seed값을.
    # 왜? -> 테스트 데이터 세트는 데이터 세트 중 랜덤으로 지정한 비율만큼 정한다.
    # 이때 랜덤으로 지정하기 때문에 seed 값을 정해주지 않는다면 학습시킬때 마다 테스트 데이터 세트가 달라지고
    # 모델이 달라지기 때문에 실습용 예제로 사용하기 위해서 seed값을 일정하게 지정해준다.
```

```python
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습 수행 -> 모델 객체 내의 fit() 함수 사용
dt_clf.fit(X_train, y_train)
```

```python
# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행
pred = dt_clf.predict(X_test)

# 예측 -> 모델 객체 내의 predict() 함수 사용
```

```python
from sklearn.metrics import accuracy_score
print('예측 정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))
print('예측 정확도 :', accuracy_score(y_test,pred)) 
# 소수점 자리수 제한할때는 format 함수를 사용할 것
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

dt_clf = DecisionTreeClassifier()
iris_data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size = 0.3, random_state=121)
```

```python
# ML 에서 데이터 세트를 train_test_split() 함수를 사용하여 학습/테스트 데이터 구분함.
# 이때 overfitting(모델이 학습 데이터에만 과도하게 최적화되어 실제 예측을 다른 데이터로 수행할 때 예측 성능이 과도하게 떨어지는 것) 발생
# 이를 해결하기 위한 방법으로 학습/검증/테스트 데이터, 세가지로 구분
# 비유를 하자면 테스트 하기전에 모의고사(검증)을 몇번 보는 것.
# 이러한 기법에는 K-fold, StratifiedKFold 존재
```

```python
# K-fold 기법
# 데이터 폴드를 K개의 fold 세트로 나누고 학습/검증을 K 번 반복하는 것
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

# 5개의 폴드 세트로 분리하는 KFold 객체와 폴드 세트별 정확도를 담을 리스트 객체 생성
kfold = KFold(n_splits = 5) # 데이터세트를 5폴더로 등분함. 4개를 학습데이터, 1개는 검증데이터로 사용
cv_accuracy = []
print('붓꽃 데이터 세트 크기 : ', features.shape[0])
```

```python
n_iter = 0

# KFold 객체의 split()를 호출하면 폴드 별 학습용, 검증용 테스트의 로우 인덱스를 array로 반환
for train_index, test_index in kfold.split(features):
    #kfold.split()으로 반환된 인덱스를 이용해 학습용, 검증용 테스트 데이터 추출
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    # 학습 및 예측
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    
    # 반복 시마다 정확도 측정
    accuracy = np.round(accuracy_score(y_test,pred),4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    
    n_iter+=1
    print('\n#{0} 교차 검증 정확도 : {1}, 학습 데이터 크기 : {2}, 검증 데이터 크기 : {3}'.format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스 : {1}'.format(n_iter, test_index))
    cv_accuracy.append(accuracy)
    
# 개별 iteration별 정확도를 합하여 평균 정확도 계산
print('\n## 평균 검증 정확도 : ',np.mean(cv_accuracy))
```
